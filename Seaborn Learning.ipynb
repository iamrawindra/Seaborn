# 1.Histogram:-

```python
from numpy.random import randn
dataset1=randn(100)
plt.hist(dataset1) #default bin size=10
```

# 2.Joint plot:-

```python
data1=randn(1000)
data2=randn(1000)
sns.jointplot(data1, data2, color='red')
```

```python
data1=randn(1000)
data2=randn(1000)
sns.jointplot(data1, data2, color='red',kind='hex')
```

# Violin plot:-

it is very similar to box plot, difference is violin plot gives you box+kde(box with distribution), which is easy to interpret.

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill', data=tips,hue='sex',split=True)
```

# strip plot:-

same as scatter plot with one variable is categorical

```python
plt.figure(figsize=(12,8))
sns.stripplot(x='day',y='total_bill',data=tips)
```

Disadvantage:-here you can not tell how many bubble are there on a single day

```python
#use:-jitter
plt.figure(figsize=(10,7))
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
#use:Hue
plt.figure(figsize=(10,7))
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True, hue='sex')
```

```python
#use split
plt.figure(figsize=(12,8))
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='day',y='total_bill',data=tips, color='red')
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
#we can use the combination of strip and violin plot as strip plot
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill',data=tips, color='blue')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill',data=tips, palette=['blue','red','green','purple'])
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill',data=tips, color='red', hue='sex')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x='day',y='total_bill',data=tips, color='red', hue='day')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

# Matrix Plot:

some times you need the visualization in matrix form during the data analysis,there are many matrix plots, fews famous are:
<br>1.heatmap</br>


it is just a color coding of row and column.
<br>-for heat map your data should always be in matrix form.</br>
<br>-both row and column should be in variable form</br>


<br>--there are two ways to make heatmap</br>
<br>1.using pivot table</br>
<br>2.using correlation data</br>

```python
#importing flight data set from seaborn
flight=sns.load_dataset('flights')
flight.head()
```

```python
tc=tips.corr()
tc
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(tc)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(tc,annot=True)
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flight.head()
```

```python
fp=flight.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fp)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fp,linecolor='white',linewidths=1)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fp,linecolor='white',linewidths=1,annot=True)
```

# Clustermap:

Hierchical Clustering to produce clustered version of heat map, compare Heatmap and clustermap figure
